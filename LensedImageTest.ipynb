Lensed image test.  Load source image and raytrace using test lens.

```python
%matplotlib inline
import evillens as evil
```

First, we create a gravitational lens.  We'll load in a kappa map also.

```python
lens = evil.GravitationalLens(0.4,1.5)
lens.read_kappa_from("analytic_SIE_kappa.fits")
lens.plot("kappa")
```

Now load an image to be the source

```python
lens.source = evil.Source(1.5)
lens.source.Read_Source_from("analytic_SIE_kappa.fits")

```

Compute the Deflection Angles

```python
lens.deflect()
```

Perform the Ray tracing

```python
lens.raytrace()
```

Plot the resulting image

```python
lens.plot('lensed image')

```

```python
lens.plot('non-lensed image')
```
