```python
%matplotlib inline
import evillens as evil
import numpy as np
```

```python
lens = evil.AnalyticSIELens(0.2,3.0)
lens.setup_grid(NX=2000,NY=2000,pixscale=0.002)
lens.build_kappa_map(sigma=200.0,centroid=[0.00,0.00], q3=0.9)
lens.deflect()
```

```python
lens.plot('kappa')
```

```python
lens.source = evil.Source(3.0)
lens.source.setup_grid(NX=2000,NY=2000,pixscale=0.0005)
#make source a gaussian with width sqrt(.0005), peak flux 1 mJy, and centroid [0.02,-0.02]
lens.source.intensity = np.exp(-0.5*(np.sqrt((lens.source.beta_x-0.03)**2+(lens.source.beta_y-0.13)**2)/0.05)**2)
```

```python
lens.raytrace()
lens.plot('lensed image')
print(np.average(lens.x))
```

```python
lens.plot('non-lensed image')
```
