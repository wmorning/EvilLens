```python
%matplotlib inline
import evillens as evil
import numpy as np
```

```python
lens = evil.AnalyticSIELens(0.4,3.0)
lens.setup_grid(NX=300,NY=300,pixscale=0.01)
lens.build_kappa_map(sigma=150.0,centroid=[0.0001,0.0001], q=0.65)
```

```python
lens.plot('kappa')
```

```python
lens.source = evil.Source(3.0)
lens.source.setup_grid(NX=2000,NY=2000,pixscale=0.001)
lens.source.intensity = np.exp(-0.5*((lens.source.beta_x-0.02)**2+(lens.source.beta_y+0.02)**2)/0.0005)
```

```python
lens.deflect()
lens.raytrace()
lens.plot('lensed image')
```

```python
lens.plot('non-lensed image')
```

```python
lens.write_image_to('./lair/DrEggman.fits')
```
